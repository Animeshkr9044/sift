```python
reviews = df[df["topic"] == "General Feedback"]["review_text"].tolist()
total_reviews = len(reviews)

one_word = len([r for r in reviews if len(r.strip().split()) == 1])
two_words = len([r for r in reviews if len(r.strip().split()) == 2]) 
three_words = len([r for r in reviews if len(r.strip().split()) == 3])
more_than_five = len([r for r in reviews if len(r.strip().split()) > 5])

print(f"Total reviews: {total_reviews}")
print(f"One word reviews: {one_word}")
print(f"Two word reviews: {two_words}")
print(f"Three word reviews: {three_words}") 
print(f"More than 5 word reviews: {more_than_five}")
```

Most reviews fall under "General Feedback" category, with many being less than 5 words that don't provide meaningful insights. These short reviews should be removed during preprocessing as they add noise without providing valuable signal for analysis.
